```python
#Removing outliers
new_data.drop(new_data[(new_data['new_cases']>1000) & (new_data['new_deaths']==0)].index.to_list(), inplace=True)
```

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
X_train_new, X_test_new, y_train_new, y_test_new = train_test_split(new_data['new_cases'], new_data['new_deaths'], test_size=0.2)
X_train = np.array(X_train_new).reshape(-1,1)
y_train = np.array(y_train_new).reshape(-1,1)
lr.fit(X_train,y_train)
```

```python
X_test = np.array(X_test_new).reshape(-1,1)
y_test = np.array(y_test_new).reshape(-1,1)
```

```python
y_pred = lr.predict(X_test)
```

```python
y_pred
```

```python
from sklearn.metrics import r2_score, classification_report, confusion_matrix
r2_score(y_test, y_pred)
```
